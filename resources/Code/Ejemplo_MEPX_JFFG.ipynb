En este ejemplo, MEPX está generando un código para multiplicar matrices 2x2 con números entre 0 y 1, esto lo hace entrenandose a partir del siguiente dataset

image.png

Para comprobar la exactitud del código generado, se van a pasar varios ejemplos de 2 matrices 2x2 (escritas como un vector de 8 elementos) generados aleatoriamente y se va a comprobar con la salida del método matmul de numpy

```python
import numpy as np


#EJECUCIÓN DE MEPX
#Generamos el vector de 8 numeros
x = np.array([np.random.random() for _ in range(8)])
#Generamos el arreglo vacío para guardar las salidas
outputs = [0] * 4
#Llamamos el código de MEPX
mepx(x, outputs)
#Imprimimos las salidas


#EJECUCIÓN DE NUMPY
#Redimensionamos el arreglos a 2 matrices 2x2
A = x[:4]
A = A.reshape(2,2)
B = x[4:]
B = B.reshape(2,2)
#Multiplicacmos la matrices
C = np.matmul(A,B)
#Imprimimos el resultado


#Impresiones
print("Para la entrada:",x)
print("El resultado de MEPX es:")
for i in range(4):
  print(outputs[i])
print("El resultado de NUMPY es:",C.reshape(4))

```

Se puede observar que el código generado por MEPX obtiene una muy buena aproximación para los elementos de la primera fila de la matriz resultante de la multiplicación, sin embargo, para la segunda fila no se acerca mucho el valor arrojado por MEPX al valor real, por lo que la multiplicación esta hecha "a medias"; esto puede deberse a la poca cantidad de datos con la que fue elaborado el programa, debido a que el dataset solo cuenta con 10 elementos, es posible, que de contar con muchos más ejemplos se pueda obtener un código que aproxime mejor la multiplicación de matrices 2x2
